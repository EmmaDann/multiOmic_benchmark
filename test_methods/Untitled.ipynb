```python
sc.pp.neighbors(adata_hvg, n_pcs=30)
```

```python
sc.tl.leiden(adata_hvg)
```

```python
sc.pl.umap(adata_hvg, color=["leiden", "annotation", "tech"], ncols=2)
```
